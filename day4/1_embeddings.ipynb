# Gensim
'''
Gensim, short for "Generate Similar," is a Python framework designed to efficiently process large volumes of text data and extract semantic topics from documents

Key features of Gensim
----------------------
** Topic Modeling: Uncover abstract topics in a corpus (uses LDA (Latent Dirichlect Allocation) algorithm)
** Word and Document Embeddings: Represent words and documents as vectors for similarity analysis
'''

```python
# pip install gensim

# import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import numpy as np
```

```python
def cosine_sim(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
docs = [
    "income tax filing was easy but the refund took long",
    "the gst portal was slow and the payment failed twice",
    "my credit card statement showed an unknown charge",
    "the bank processed my loan application quickly",
    "interest rates increased and emi became higher this month",
    "the auditor requested additional documents for verification",
    "received a notice for mismatch in tds records",
    "the mutual fund dashboard is clean and easy to track",
    "my demat account login failed due to server issues",
    "the tax consultant explained deductions very clearly",
    "online banking was down during peak hours",
    "the investment app recommended low risk funds",
    "my upi transaction was debited but not credited",
    "the insurance claim was approved without much delay",
    "the ca helped me compute capital gains accurately",
    "the pension contribution details were not updated",
    "the trading platform froze during market volatility",
    "the bank resolved my kyc issue within minutes",
    "the tax calculator showed incorrect slab values",
    "my sip payment bounced due to mandate failure"
]
```

```python
# get the unique words from the doc
reviews = []

# split each document into individual words
for d in docs:
    words = d.split()
    reviews.append(words)
print(reviews)
```

```python
# Build the Word 2 Vector model

'''
vector_size → embedding length
window      → context understanding
min_count   → remove rare words
sg          → CBOW vs SkipGram
epochs      → training rounds
alpha       -> training parameter
'''
model = Word2Vec(reviews, vector_size=50, window=3, min_count=1)
print(model)
```

```python
# unique words in the reviews
unique_words = list(model.wv.index_to_key)
print(unique_words)
```

```python
word1 = "payment"; word2 = "pension"

A = model.wv[word1]
B = model.wv[word2]

print(f"Vector representation of Word '{word1}' \n{A}\nShape = {A.shape}")
print()
print(f"Vector representation of Word '{word2}' \n{B}\nShape = {B.shape}")
```

# Sentence Embeddings
# to calculate Sentence Embeddings, 
#   a) split sentence into words
#   b) take average

```python
def sentence_embedding(tokens, model):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)
```

```python
# each sentence in docs is converted into words
tokenized = [simple_preprocess(doc) for doc in docs]

# call the Sentence embeddings function
sent_embeddings = [sentence_embedding(tokens, model) for tokens in tokenized]

# print the total sentence embeddings. they are same as the length of the documents
len(sent_embeddings)

# print the embeddings of the first sentence
sent_embeddings[0]
```

```python
# Sample sentence similarity
# --------------------------

query = ["customer is yet to complete kyc"]

q_token = [simple_preprocess(doc) for doc in query]
# print(q_token)

q_emb = [sentence_embedding(tokens, model) for tokens in q_token]
```

```python
results = {}

for i in range(len(sent_embeddings)):
    a_emb = sent_embeddings[i]
    results[i] = cosine_sim(q_emb,a_emb)
```

```python
sorted_results = sorted(results.items(),key=lambda x:x[1],reverse=True)
print(sorted_results[:2])
```
