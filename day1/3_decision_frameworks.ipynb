# Three distinct ways of executing work
3.1 Deterministic systems
3.2 RPA (Robotic Process Automation) represent
3.3 Autonomous systems

```python
import math, pandas as pd, time
```

```python
# 1) Deterministic Systems : Always the same result

radius = 2.3

for i in range(1,26):
    print(f"{i} Area of the circle is {round(math.pi * math.pow(radius,2),1) }")
```

```python
# 2) RPA

# i)    Reads data from a source.
# ii)   Applies predefined rules.
# iii)  Takes action automatically.
# iv)   Produces an output report.
# v)    No reasoning, planning, or learning involved.

def financedata():
    data = pd.read_csv("invoice.csv")
    return(data)
```

```python
# Set the following invoice id statuses to approved if the amount is more than 15000. Else set status as on-hold

def run(s=3):
    data = financedata()
    sample_data = data.sample(s)

    selected_invoices = sample_data.invoice_id.tolist()

    for ndx, record in sample_data.iterrows():
        inv_id = record.loc["invoice_id"]
        amt = record.loc["amount"]

        if amt > 5000:
            data.loc[ndx,"status"] = "Approved"

    data.to_csv("invoice.csv",index=False)
    
    data = financedata()
    print(data[data.invoice_id.isin(selected_invoices)].sort_values("amount",ascending=False))
```

```python
run(5)
```

```python
# 3) Autonomous System

# i)    Has a goal.
# ii)   Evaluates the situation.
# iii)  Chooses an action.
# iv)   Can adapt its behavior.


# check a file every 10 seconds. 
# update the data as follows:
# if amount > 5000, set status = Approved
# if amount between 3000-4999, set status = Approved with Conditions
```

```python
def run2(s=3):
    data = financedata()
    sample_data = data.sample(s)

    selected_invoices = sample_data.invoice_id.tolist()

    for ndx, record in sample_data.iterrows():
        inv_id = record.loc["invoice_id"]
        amt = record.loc["amount"]

        if amt > 5000:
            data.loc[ndx,"status"] = "Approved"
        elif amt > 3000:
            data.loc[ndx,"status"] = "Approved with Condition"
        else:
            pass

    data.to_csv("invoice.csv",index=False)
    
    data = financedata()
    print(data[data.invoice_id.isin(selected_invoices)].sort_values("amount",ascending=False))
```

```python
try:
    print("This Autonomous agent runs every 10 seconds. Press CTRL+C to stop the process")
    while True:
        run(3)
        time.sleep(5)

except KeyboardInterrupt:
    print("Agent stopped by user")
```
